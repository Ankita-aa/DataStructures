1. What are data structures, and why are they important?
- Data structures are ways of organizing and storing data efficiently. They are important because they determine how data can be accessed, modified, and stored, which impacts the performance of algorithms. Examples include lists, tuples, sets, and dictionaries.

2. Difference between mutable and immutable data types
- Mutable: Can be changed after creation. Example: list
- Immutable: Cannot be changed after creation. Example: tuple


3. Differences between lists and tuples
- Feature	List (list)	Tuple (tuple)
- Mutable?	list are mutable and tuples are not.
- Performance : list are Slower (modifiable) while tuple are Faster (immutable)
- Syntax	[] (square brackets for list )	() (parentheses for tuple)
4. How dictionaries store data
- Dictionaries store data as key-value pairs using hashing, allowing fast lookups.


5. Why use a set instead of a list?
- Removes duplicates automatically

6. What is a string, and how is it different from a list?
- A string is an immutable sequence of characters.

- A  list is mutable and can store multiple data types.

7. How do tuples ensure data integrity?
- Since tuples are immutable, their contents cannot be changed, ensuring data integrity.


8. What is a hash table, and how does it relate to dictionaries?
- A hash table stores data in key-value pairs using hash functions for fast lookups and Python dictionaries are implemented using hash tables.

9. Can lists contain different data types?
- Yes! Lists can hold multiple data types:

10. Why are strings immutable?
- Strings are stored in memory as sequences of characters. Immutability prevents accidental modifications and allows memory optimization.

11. Advantages of dictionaries over lists
- Faster lookups (O(1)) compared to lists (O(n))

- Key-value pairs make data retrieval easier.

12. When is a tuple preferable over a list?
- When data should not be changed (e.g., database records, coordinates).

- Faster performance than lists.
