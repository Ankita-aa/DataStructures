```python
# Q1: What is NumPy, and why is it widely used in Python?
# A: NumPy is a powerful numerical computing library in Python. It's widely used for handling large, multi-dimensional arrays and matrices, and provides functions for performing mathematical operations efficiently.

# Q2: How does broadcasting work in NumPy?
# A: Broadcasting is a technique that allows NumPy to perform arithmetic operations on arrays of different shapes. Smaller arrays are "broadcast" over larger arrays so they have compatible shapes.

# Q3: What is a Pandas DataFrame?
# A: A Pandas DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It’s like a table in a database or an Excel spreadsheet.

# Q4: Explain the use of the groupby() method in Pandas.
# A: The `groupby()` method is used to group data based on a column or condition and apply aggregation functions like sum, mean, or count to each group.

# Q5: Why is Seaborn preferred for statistical visualizations?
# A: Seaborn is built on top of Matplotlib and provides a high-level API for drawing attractive and informative statistical graphics with less code and better default styles.

# Q6: What are the differences between NumPy arrays and Python lists?
# A: NumPy arrays are more memory efficient, allow vectorized operations, and support multi-dimensional data, while Python lists are general-purpose containers without built-in mathematical capabilities.

# Q7: What is a heatmap, and when should it be used?
# A: A heatmap is a data visualization technique that shows the magnitude of a phenomenon as color in two dimensions. It’s used to represent correlation matrices or frequency of occurrence.

# Q8: What does the term “vectorized operation” mean in NumPy?
# A: Vectorized operations allow performing operations on entire arrays without explicit loops, leading to faster and more concise code.

# Q9: How does Matplotlib differ from Plotly?
# A: Matplotlib creates static, publication-quality plots, while Plotly is used for interactive, web-based visualizations with zooming, hovering, and click capabilities.

# Q10: What is the significance of hierarchical indexing in Pandas?
# A: Hierarchical indexing allows multiple index levels on an axis, enabling more complex data representations and easier reshaping or slicing of DataFrames.

# Q11: What is the role of Seaborn’s pairplot() function?
# A: The `pairplot()` function visualizes pairwise relationships in a dataset by plotting scatterplots and histograms for each variable combination, helpful for EDA.

# Q12: What is the purpose of the describe() function in Pandas?
# A: The `describe()` function provides summary statistics like mean, std, min, max, and quartiles for each numeric column in a DataFrame.

# Q13: Why is handling missing data important in Pandas?
# A: Handling missing data is crucial for maintaining data quality and ensuring accurate analysis, as missing values can skew results or break operations.

# Q14: What are the benefits of using Plotly for data visualization?
# A: Plotly offers interactive, zoomable, and visually appealing plots that are great for web dashboards and exploratory analysis with minimal setup.

# Q15: How does NumPy handle multidimensional arrays?
# A: NumPy handles multidimensional arrays using the `ndarray` object, which allows fast and efficient manipulation and mathematical operations over any number of dimensions.

# Q16: What is the role of Bokeh in data visualization?
# A: Bokeh is a Python library for creating interactive, browser-based visualizations with high performance for large or streaming datasets.

# Q17: Explain the difference between apply() and map() in Pandas.
# A: `map()` is used for element-wise operations on Series, while `apply()` can be used on both Series and DataFrames for row/column-wise operations.

# Q18: What are some advanced features of NumPy?
# A: Advanced features of NumPy include broadcasting, vectorization, masked arrays, memory-mapped files, random sampling, and linear algebra operations.

# Q19: How does Pandas simplify time series analysis?
# A: Pandas provides date/time indexing, resampling, frequency conversion, and time zone handling, which simplify time-based data analysis and manipulation.

# Q20: What is the role of a pivot table in Pandas?
# A: Pivot tables are used to reshape and summarize data, allowing aggregation of values across multiple dimensions for better insight.

# Q21: Why is NumPy’s array slicing faster than Python’s list slicing?
# A: NumPy arrays are stored in contiguous memory blocks, enabling faster access and manipulation compared to Python lists which store references.

# Q22: What are some common use cases for Seaborn?
# A: Common use cases include visualizing distributions (hist, kde), relationships (scatter, pairplot), and categorical data (box, violin, bar plots), especially for statistical analysis.
```

```python
#4
import pandas as pd
import seaborn as sns

df = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [4, 3, 2, 1],
    "C": [2, 3, 4, 5]
})

cor = df.corr()
sns.heatmap(cor, annot = True, cmap = "coolwarm")
plt.show()
```

```python
#5
import plotly.express as px

data = {"Fruits": ["Apples", "Bananas", "Cherries"], "Count": [10, 15, 7]}
df = pd.DataFrame(data)

fig = px.bar(df, x="Fruits", y="Count", title="Fruit Count")
fig.show()
```
