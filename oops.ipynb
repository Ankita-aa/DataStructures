OOPS


1 What is Object-Oriented Programming (OOP)?
> oops is programming paradigm based on the concept of 'object' which can contain data that is attribute and methods. It allows for modeling real world entities using classses and object

2 What is a class in OOP?
> class is a blueprint or templete for creating objects.

3 What is an Object in OOP?
> Object is the instance of the class. It has actuall values for the properties defined by the class and can use its method.

4 What is the difference between abstraction and encapsulation?
> Abstraction hides complex implementation details and shows only relevant features (focuses on what to do).
> Encapsulation hides the internal state of the object and restricts direct access (focuses on how to protect data).


5 What are dunder methods in Python?
 > Dunder methods are built in classes or methods in python for creating custom objects and also use to implement overloading in python


6 Explain the concept of inheritance in OOP?
> Inheritance allows one class (child/derived) to inherit attributes and methods from another class (parent/base). It promotes code reuse.

7 What is polymorphism in OOP?
> Polymorphism allows objects of different classes to be treated as objects of a common superclass. Methods with the same name can behave differently depending on the object.

8 How is encapsulation achieved in Python?
> Encapsulation is achieved by:  
Using private variables (prefix with _ or __)   
Providing getter/setter methods to control access

9 What is a constructor in Python?
>A constructor is a special method called automatically when an object is created. In Python, it’s defined using  __ init __().

10 What are class and static methods in Python?
> Class Method: Uses @classmethod decorator. It takes cls as the first argument and can access or modify class state. Static Method: Uses @staticmethod decorator. It doesn’t take self or cls and behaves like a regular function within a class.

11 What is method overloading in Python?
> Python does not support traditional method overloading like Java or C++. Instead, you can simulate it using default arguments

12 What is method overriding in OOP+?
> Method overriding is when a child class defines a method with the same name as a method in the parent class to provide a different implementation.

13 What is a property decorator in Python?
> @property allows you to define a method as a getter (like accessing an attribute). You can also define @<property>.setter.

14 Why is polymorphism important in OOP+?
> Polymorphism increases flexibility and scalability by allowing different objects to be used interchangeably if they implement the same interface or method.
Example: You can call .draw() on different shape classes (Circle, Square) without knowing the exact type.


15 What is an abstract class in Python?
> An abstract class cannot be instantiated directly and may contain abstract methods (declared but not implemented). Use the abc module.

16 What are the advantages of OOP?
> Modularity: Code is organized into objects, Reusability: Classes can be reused and extended, Scalability: Easy to manage and scale code, Maintainability: Cleaner and easier to debug.


17 What is the difference between a class variable and an instance variable?
> Class Variable: Shared among all instances whereas Instance Variable: Unique to each object.

18 What is multiple inheritance in Python?
> A class can inherit from more than one parent class, allowing access to features of all parents.

19 Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python?
> Both __str__ and __repr__ are dunder (double underscore) methods used to define how an object should be represented as a string.

20 H What is the significance of the ‘super()’ function in Python?
> super() is used to call a method from the parent class. It’s especially helpful in inheritance to reuse the parent class’s methods and avoid repeating code.




21 What is the significance of the __del__ method in Python?
>__del__() is the destructor method in Python. It's automatically called when an object is about to be destroyed.

22 What is the difference between @staticmethod and @classmethod in Python?
> @staticmethod	takes No self or cls, Cannot access or modify class	whereas @classmethod takes cls as the first argument, Can access/modify class data.


23 How does polymorphism work in Python with inheritance?
> Polymorphism allows you to use the same method name across different classes, and Python decides which implementation to call based on the object.

24 What is method chaining in Python OOP?
>Method chaining allows multiple methods to be called in a single line, one after the other. This works only if each method returns self.


25 What is the purpose of the __call__ method in Python?
> If a class has a __call__() method, its instances behave like a function — you can call the object directly.

```python
#2

import abc

class FindArea:
  @abc.abstractmethod
  def area(self):
    pass

class Square(FindArea):
  def __init__(self, side):
        self.side = side

  def area(self):
    return self.side*self.side

class Rectangle(FindArea):

  def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

  def area(self):
    return self.length*self.breadth

s1 = Square(5);
print(s1.area())

r1 = Rectangle(2,3);
print(r1.area())
```

```python
#5

class Bank:
  def __init__(self, balance):
    self.__balance = balance

  def __check_balance(self):
    return self.__balance

  def __deposit(self, amount):
    self.__balance = self.__balance + amount

  def __withdrawal(self, withdrawAmount):
    if(self.__balance>= withdrawAmount):
      self.__balance = self.__balance - withdrawAmount

cust = Bank(500)
cust._Bank__check_balance()
```

```python
##10

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        # Add corresponding components
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Create two vectors
v1 = Vector(2, 3)
v2 = Vector(4, 1)

# Add them using +
v3 = v1 + v2

print("Result:", v3)
```

```python
#14
class Employee:

  def __init__(self,hours,hourate):
    self.hours = hours
    self.hourate = hourate


  def calculated_salary(self):
      return self.hours*self.hourate*30*12


class Manager(Employee):
  def __init__(self,hours,hourate,bonus):
    super().__init__(hours,hourate)
    self.bonus = bonus

  def add_bonus(self):
    return self.bonus+ self.calculated_salary()

e1 = Employee(8,1000)
e1.calculated_salary()

m1 = Manager(8,1000,10000)
m1.add_bonus()
```

```python
#16
import abc

class Animal:
  @abc.abstractmethod
  def sound(self):
    pass

class Cow(Animal):
  def __init__(self,para):
    self.para = para
  def sound(self):
    print("can hear cow",self.para)

class Sheep(Animal):
  def sound(self):
    print("can hear sheep")

c1 = Cow(5)
print(c1.sound())

s1 = Sheep()
print(s1.sound())
```
