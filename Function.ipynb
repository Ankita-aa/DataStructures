Python allows defining and calling functions in multiple ways:

- Standard Function

6. Explain the concept of generators in Python and how they are defined.
- A generator is a special function that uses yield instead of return. It remembers its state and resumes execution when next() is called.

def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

gen = count_up_to(3)
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
7. What are the advantages of using generators over regular functions?
- Memory-efficient (Does not store values in memory).
- Faster execution (Lazy evaluation).
- Useful for handling large datasets.

```python
#5
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0

    def __iter__(self):
        return self  # An iterator must return itself

    def __next__(self):
        if self.count < self.n:
            fib_number = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return fib_number
        else:
            raise StopIteration

# Example usage
fib = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers
for num in fib:
    print(num, end=" ")
```

```python
#6
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i  # Generate 2^i


for power in powers_of_two(5):  # Generates 2^0 to 2^5
    print(power, end=" ")

```

```python
#7
def read_file_line_by_line(filename):
    """Generator function to read a file line by line."""
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()  # Yield each line, removing extra spaces/newlines

# Example usage
file_path = "example.txt"
for line in read_file_line_by_line(file_path):
    print(line)
```
